```python
import numpy as np
import pandas as pd
```

```python
datos = pd.read_csv('Bienestar.csv', encoding='latin1', low_memory=False)
datos.head()
```

```python
datos.drop(['anio_reg', 'trimestre', 'hog_ent_1', 'hog_ent_2', 'id_pobla', 'sexo', 'edad', 'edo_conyug',
            'grado_inst', 'nivel_inst', 'scian', 'sinco', 'clasif_emp','remod', 'maltra_e', 'otranec_e'], axis=1, inplace=True)
datos.head()
```

```python
datos.describe()
```

```python
# Se reemplazan valores NaN por ceros
datos.fillna(0, inplace=True)
```

```python
from sklearn import preprocessing
def encode_features(datos):
    features = list(datos.drop(['folio', 'gr_satis', 'gr_feliz'], axis=1).columns)
    #
    for feature in features:
        le = preprocessing.LabelEncoder()
        le.fit(datos[feature])
        datos[feature] = le.transform(datos[feature])
    #
    return datos
```

```python
d_encode = encode_features(datos)
d_encode.head()
```

```python
# Modelo para la variable de salida 'gr_satis'
# Separando en training
from sklearn.model_selection import train_test_split
#Borramos las columnas folio, gr_satis, gr_feliz
X_all = d_encode.drop(['folio', 'gr_satis', 'gr_feliz'], axis=1)
y_all = d_encode['gr_satis']
```

```python
num_test = 0.20 # aquÃ­ se estÃ¡ especificando 20% para test
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size=num_test)
```

```python
# Seleccionamos el tipo de clasificador, en este caso RandomForest. 
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(1000)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import make_scorer, accuracy_score, f1_score
predictions = clf.predict(X_test)
# se imprime el accuracy
print("Accuracy: ", round(accuracy_score(y_test, predictions) * 100, 2))
#Imprimimos la f-score
# Opciones average = 'micro', 'macro', 'weighted'
print("F1-Score: ", round(f1_score(y_test, predictions, average='weighted') * 100, 2))
```

```python
# Matriz de confusion (falsos positivos, falsos negativos)
# https://towardsdatascience.com/understanding-confusion-matrix-a9ad42dcfd62
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X_train, y_train, cv=6)
print(scores)
```

```python
#Podemos cambiar la mÃ©trica de evaluaciÃ³n con el parÃ¡metro scoring
scores = cross_val_score(clf, X_test, y_test, cv=6, scoring='f1_weighted')
print(scores)
```
